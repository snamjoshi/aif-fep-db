```python
import arxiv
import pandas as pd

from biorxiv_retriever import BiorxivRetriever
```

## ArXiv

```python
def arxiv_download(search_term):
    search = arxiv.Search(
        query = search_term,
    )

    entry_id = []
    updated = []
    published = []
    title = []
    author = []
    categories = []

    for result in arxiv.Client().results(search):
        entry_id.append(result.entry_id)
        updated.append(result.updated)
        published.append(result.published)
        title.append(result.title)
        author.append(result.authors)
        categories.append(result.categories)
        
    arxiv_db = pd.DataFrame({
        "id": entry_id,
        "updated": updated,
        "published": published,
        "title": title,
        "authors": author,
        "categories": categories
    })

    return arxiv_db

arxiv_active_inference = arxiv_download("active inference")
arxiv_karl_friston = arxiv_download("bayesian mechanics")
arxiv_friston = arxiv_download("friston")
arxiv_free_energy = arxiv_download("free energy principle")
arxiv_predictive_coding = arxiv_download("predictive coding")
arxiv_predictive_processing = arxiv_download("predictive processing")

arxiv_db = pd.concat([arxiv_active_inference, arxiv_karl_friston, arxiv_friston, arxiv_free_energy, arxiv_predictive_coding, arxiv_predictive_processing])

arxiv_db = arxiv_db.reset_index()
```
